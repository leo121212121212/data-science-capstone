<H1>Leo’s quest for the best (NYC journey for good vet)</H1>
<h2>Description of the problem and a discussion of the background</h2>
<p>I will visit NY city this summer for a business trip. I will take with me my beloved mini pig Sammy. I adore Sammy, but I am afraid that because of stress and jet lag Sammy may catch a cold or some kind of “mal de mer”. So, I need to make a list of vet clinics in the areas (boroughs) of NYC and put them on a map to help me to identify the best place for a room rent, where I will stay for the trip.  So, this is the problem.</p>
<p>We (Sammy and I) are the main stakeholders of this research. Sammys life may be in danger, so I must be careful doing this work</p>

<h2> A description of the data and how it will be used to solve the problem. </h2>
<p>I will use Foursquare API to get all the info.</p>
<p>This is the sequence of actions:</p>
<ol>
<li>Get the coordinates of NYC centre. According to <a href="https://www.latlong.net">https://www.latlong.net</a> the latitude of New York City, NY, USA is 40.730610, and the longitude is -73.935242. </li>
<li>According to Wikipedia (<a href="https://en.wikipedia.org/wiki/New_York_City#Geography">https://en.wikipedia.org/wiki/New_York_City#Geography</a>) the area of NYC is 1213.37 square km, so if we assume that 1213.37 is the square of the circle, we can easily count its radius - 8.921 km. Of course NYC is not circular by its shape, so we will need to make sure that vet clinics are situated in NYC, but not in New Jersey or other places.</li>
<li>Using Foursquare API we will get (collect) the data frame of vet clinics.</li>
<li>Then I will prepare the data for the research (get rid of unnecessary or irrelevant data, columns etc.)</li>
<li>Then I will put all the clinics on the map and will obtain a list of hotels near the place.</li>
<li>The last step is to choose the hotel and book it for my trip.</li>
</ol>
<p>So, here is the plan!</p>
